```python

from nltk import bigrams, trigrams, FreqDist
# Calculate biagram frequency
biagram_freq = FreqDist(bigrams(C))
# Calculate trigram frequency
trigram_freq = FreqDist(trigrams(C))

# Convert to dictionary for easier plotting
biagram_freq_dict = {" ".join(k): v for k, v in biagram_freq.items()}
trigram_freq_dict = {" ".join(k): v for k, v in trigram_freq.items()}

# Select top 10 bigrams and trigrams
top_biagrams = dict(sorted(biagram_freq_dict.items(), key=lambda item: item[1], reverse=True)[:20])
top_trigrams = dict(sorted(trigram_freq_dict.items(), key=lambda item: item[1], reverse=True)[:10])

# Plotting bigrams
plt.figure(figsize=(14, 7))
plt.bar(top_biagrams.keys(), top_biagrams.values())
plt.xlabel('Bigrams')
plt.ylabel('Frequency')
plt.title('Top 10 Bigrams Frequency Distribution')
plt.xticks(rotation=45)
plt.show()

# Plotting trigrams
plt.figure(figsize=(14, 7))
plt.bar(top_trigrams.keys(), top_trigrams.values())
plt.xlabel('Trigrams')
plt.ylabel('Frequency')
plt.title('Top 10 Trigrams Frequency Distribution')
plt.xticks(rotation=45)
plt.show()

```

```python
from nltk import ngrams, FreqDist
import matplotlib.pyplot as plt

# Function to plot n-gram frequency distribution
def plot_ngram_frequency(text, n, top_n=10):
    # Calculate n-gram frequency
    ngram_freq = FreqDist(ngrams(text, n))

    # Convert to dictionary for easier plotting
    ngram_freq_dict = {" ".join(k): v for k, v in ngram_freq.items()}

    # Select top n-grams
    top_ngrams = dict(sorted(ngram_freq_dict.items(), key=lambda item: item[1], reverse=True)[:top_n])

    # Plotting
    plt.figure(figsize=(14, 7))
    plt.bar(top_ngrams.keys(), top_ngrams.values())
    plt.xlabel(f'{n}-grams')
    plt.ylabel('Frequency')
    plt.title(f'Top {top_n} {n}-grams Frequency Distribution')
    plt.xticks(rotation=45)
    plt.show()

# Plot for 4-gram, 5-gram, and 6-gram
plot_ngram_frequency(C, 2)
plot_ngram_frequency(C, 3)
plot_ngram_frequency(C, 4)
plot_ngram_frequency(C, 5)
plot_ngram_frequency(C, 6)
plot_ngram_frequency(C, 7)
plot_ngram_frequency(C, 8)
plot_ngram_frequency(C, 9)
plot_ngram_frequency(C, 10)

```
