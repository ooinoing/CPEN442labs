### Letter Frequency

```python
import matplotlib.pyplot as plt
from collections import Counter

def visualize_letter_frequency(text, color='skyblue', title='Letter Frequency Distribution'):
   
    frequency = Counter(text)
    sorted_frequency = {k: v for k, v in sorted(frequency.items(), key=lambda item: item[1], reverse=True) if k.isalpha()}
    
    keys = sorted_frequency.keys()
    values = sorted_frequency.values()
    
    # Plotting
    plt.figure(figsize=(10, 5))
    plt.bar(keys, values, color=color)
    plt.xlabel('Letters')
    plt.ylabel('Frequency')
    plt.title(title)
    plt.show()

visualize_letter_frequency(C, "salmon","Letter Frequency of CipherText" )
visualize_letter_frequency(P, "skyblue","Letter Frequency of PlainText" )
```

### Diagraph Frequency

```python
def calculate_digraph_frequency(text):
    text = text.replace(" ", "").lower()
    digraphs = [text[i:i+2] for i in range(0,len(text) - 1,  2)]
    frequency = Counter(digraphs)
    
    return frequency

def visualize_digraph_frequency(frequency, title="Frequency of Digraphs", color='skyblue'):
    # Sort digraphs by frequency in descending order
    digraphs = sorted(frequency, key=frequency.get, reverse=True)[:20]
    frequencies = [frequency[digraph] for digraph in digraphs]
    
    plt.figure(figsize=(20, 8))
    plt.bar(digraphs, frequencies, color=color)
    plt.xlabel('Digraphs')
    plt.ylabel('Frequency')
    plt.title(title)
    plt.xticks(rotation=45)
    plt.show()
    
    #return digraphs

cipher_digraph = calculate_digraph_frequency(C)
plain_digraph = calculate_digraph_frequency(P)


visualize_digraph_frequency(cipher_digraph, 'Cipher Diagraph', "salmon")
visualize_digraph_frequency(plain_digraph, 'Plaintext Diagraph')
```

### Frequently Translated Pairs

```python
from collections import Counter
import matplotlib.pyplot as plt

def collect_translated_pairs(plaintext, ciphertext, pair="th"):
    
    plaintext = plaintext.replace(" ", "").lower()
    ciphertext = ciphertext.replace(" ", "").lower()

    translated_pairs = []
    for i in range(0, len(plaintext)-1):
        if plaintext[i:i+2] == pair:
            translated_pair = ciphertext[i:i+2]
            translated_pairs.append(translated_pair)
    
    return Counter(translated_pairs)

def visualize_translated_pairs(counter,pair="th" ):
    """
    Visualizes the frequency of translated pairs.
    """
    pairs = [pair for pair, count in counter.most_common(20)]
    counts = [count for pair, count in counter.most_common(20)]
    
    plt.figure(figsize=(10, 5))
    plt.bar(pairs, counts, color='salmon')
    plt.xlabel('Translated Pairs')
    plt.ylabel('Frequency')
    plt.title(f'Frequency of Translated {pair} Pairs')
    plt.xticks(rotation=45)
    plt.show()

pairs = ["th","ma","he","om","ot"]

for pair in pairs:
    translated_pairs_counter = collect_translated_pairs(P, C, pair)
    visualize_translated_pairs(translated_pairs_counter, pair)
    
```
